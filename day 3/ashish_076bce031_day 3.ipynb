```python
#Q no 2
import random as rn
import matplotlib.pyplot as plt
import sympy as sp
import numpy as np
R='Red'
W='White'
K='Black'
B='Blue'
bag=[R,R,R,R,  W,W,W,  B,B,B,B,B,B,B,B, K,K,K,K]
# creating results array to store results
results = []
for i in range(20):
    results.append( rn.choice(bag) )
#display (results) # if you want to display results

# calculating counts to obtain probability for each color
countR=results.count(R)
countB=results.count(B)
countW=results.count(W)
countK=results.count(K)
# color array is declared for x axis of bar(histogram)
color=[R,B,W,K]
# probability array has probability of each color.. Note probability and color arrays  have color of ball in same order
probability=[countR/20,countB/20,countW/20,countK/20]


# display probability of white colored ball
print("probability of white colored ball is " , probability[2])


# plotting bar
plt.bar(color, probability)
plt.xlabel('Color of Ball')
plt.ylabel('Probability')


```

```python
# Q No 3
x=sp.symbols('x')
import math as mt
y1=np.array([])# for sine curve

y2=np.array([])# for cosine curve

y3=np.array([])# for sine scatter points

y4=np.array([])# for cosine scatter points

x_1=np.linspace(-mt.pi,2*mt.pi,100)# for curve

x_2=np.arange(0,mt.pi,mt.pi/12)# for scatter points

#filling array y1
for x_i in x_1:
    y_i=mt.sin(x_i)
    y1=np.append(y1,y_i)
#filling array y3   
for x_i in x_2:
    y_i=mt.sin(x_i)
    y3=np.append(y3,y_i)
#filling array y2    
for x_i in x_1:
    y_i=mt.cos(x_i)
    y2=np.append(y2,y_i)
#filling array y4
for x_i in x_2:
    y_i=mt.cos(x_i)
    y4=np.append(y4,y_i)
#plotting part
# sine curve
plt.plot(x_1,y1,linestyle="--")
# sine scatter points
plt.scatter(x_2,y3,label="sin(\u03F4)")
# cosine curve
plt.plot(x_1,y2,linestyle="--")
# cosine scatter points
plt.scatter(x_2,y4,label="cos(\u03F4)")
#adding labels ,grids and legends
plt.ylabel('f(x)')
plt.xlabel('\u03F4')
plt.grid(linestyle="dotted")
plt.axis([-4,7,-1.2,1.2])
plt.legend()
plt.show()
```

```python
#Qno4

#Symbol Definition
Vx,Mx,Ry,Rx,w,RAy,RAx,REy,P1,P2 = sp.symbols("V_x,M_x,R_y,R_x,w,R_A_y,R_A_x,R_E_y,P_1,P_2")


#Writing equilibrium Equations


#Sum of vertical force equals zero
print('Using Equations of Equilibrium, we have ') 
display(sp.Eq(Ry,0),sp.Eq(RAy+REy-P1-P2-w*2,0))
eq1=RAy+REy-P1-P2-w*2
print('Putting given values, we get')
eq1=eq1.subs([(P1,6),(P2,4),(w,1)])
display(sp.Eq(eq1,0))


#Sum of Horizontal forces equals zero
display('And From',sp.Eq(Rx,0),'we get',sp.Eq(RAx,0))


# sum of moment about any point (here A) equals zero
display("For equilibrium of moment about A ",sp.Eq(Mx,0),"or",sp.Eq(P1*1+P2*3+2*w*4-REy*6,0))
eq2=P1*1+P2*3+2*w*4-REy*6
eq2=eq2.subs([(P1,6),(P2,4),(w,1)])
display(sp.Eq(eq2,0))


# solving equations to obtain reation forces
values=sp.solve([eq1,eq2],[RAy,REy])
display('Solving equations , we obtain support reactions as',sp.Eq(RAy,values[RAy]),sp.Eq(REy,values[REy]),sp.Eq(RAx,0))


#calculating shear forces 
x,y=sp.symbols("x,y")
def sf(x,y,z):
    return sp.SingularityFunction(x,y,z)
sh=values[RAy]*sf(x,0,0)+(-P1)*sf(x,1,0)+(-P2)*sf(x,3,0)+(-w)*sf(x,3,1)+(w)*sf(x,5,1)+values[REy]*sf(x,6,0)
print('Calculation of Shear Force')
display(sp.Eq(Vx,sh),'After putting Respective Values')
sh=sh.subs([(P1,6),(P2,4),(w,1)])
display(sp.Eq(Vx,sh))


# calculating bending moments
momfn=sp.Integral(sh,(x)).doit()
print('Calculation of Bending Moment:')
display(sp.Eq(Mx,momfn))



# loading conditions  diagram
fig,axes=plt.subplots(nrows=1, ncols=1, sharex=True, figsize=(6, 4))
axes.plot([0,6],[0,0], linewidth=1, color="k")
axes.plot([3,5],[1,1], linewidth=1, color="k")
axes.set_title("Loading Conditions")
axes.set_ylabel("P, KN")
axes.grid( linestyle = "dotted")
#R_A_y arrow
axes.arrow(0, -23/3, 0, 23/3 - .2,
    linewidth = 2, head_length = .2, head_width = .1, color = "k")
#6kn arrow
axes.arrow(1, 6, 0,-6 +.2,
    linewidth = 2, head_length = .2, head_width = .1, color = "k")
#4kn arrow
axes.arrow(3, 4, 0,-4 +.2,
    linewidth = 2, head_length = .2, head_width = .1, color = "k")
# UDL arrows
for i in range(5):
    axes.arrow(3+0.4*(i+1), 1,0,-(1) +.2,
    linewidth = 2, head_length = .2, head_width = .1, color = "k")
#R_E_y arrow
axes.arrow(6, -13/3, 0, 13/3 - .2,
    linewidth = 2, head_length = .2, head_width = .1, color = "k")
plt.show()



# shear force diagram
print('Shear force Diagram:')
x_val=np.linspace(0,6,100)
sf=np.array([])
for x_i in x_val:
    y_i=sh.subs(x,x_i)
    sf=np.append(sf,y_i)
plt.plot(x_val,sf,color="k",linewidth=2)
plt.ylabel('V(x),KN')
plt.title('Shear Force Diagram(SFD)')
plt.grid(linestyle="dotted")
plt.show()




#bending moment digram
bm=np.array([])
x_val1=np.linspace(0,6,1000)
for x_i in x_val1:
    y_i=momfn.subs(x,x_i)
    bm=np.append(bm,y_i)
print('Bending Moment Diagram:')
plt.plot(x_val1,bm,color="k",linewidth=2)
plt.ylabel('M(x),KNm')
plt.title('Bending Moment Diagram(BMD)')
plt.grid(linestyle="dotted")
plt.show()
```
